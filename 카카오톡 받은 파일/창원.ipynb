```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

def read_dem_file(file_path):
    with open(file_path, 'r') as file:
        dem_data = []
        for line in file:
            line_data = line.strip().split()
            dem_data.append([float(value) for value in line_data])
        return np.array(dem_data)

def plot_dem_on_map(dem_data, lat_start, lat_end, lon_start, lon_end):
    m = Basemap(projection='merc', llcrnrlat=lat_start, urcrnrlat=lat_end,
                llcrnrlon=lon_start, urcrnrlon=lon_end, resolution='i')
    lons = np.linspace(lon_start, lon_end, dem_data.shape[1])
    lats = np.linspace(lat_start, lat_end, dem_data.shape[0])
    lons, lats = np.meshgrid(lons, lats)
    x, y = m(lons, lats)
    cs = m.contourf(x, y, dem_data, cmap='terrain')
    m.colorbar(cs, location='bottom', pad="10%")
    m.drawcoastlines()
    m.drawcountries()
    m.drawstates()
    plt.show()

# 파일 경로와 좌표 범위 설정
file_path = '/content/창원시 마산합포구.txt'
#lat_start, lat_end = 35, 35.9  # 위도 범위 설정
#lon_start, lon_end = 128.3, 128.9  # 경도 범위 설정
lat_start, lat_end = 0, 60  # 위도 범위 설정
lon_start, lon_end = 0, 180  # 경도 범위 설정

# DEM 데이터 읽어오기
dem_data = read_dem_file(file_path)

# DEM 데이터 지도에 나타내기
plot_dem_on_map(dem_data, lat_start, lat_end, lon_start, lon_end)
```
