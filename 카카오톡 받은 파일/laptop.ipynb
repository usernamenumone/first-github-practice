```python
import pandas as pd
import matplotlib.pyplot as plt

# 업로드한 파일 읽기
df = pd.read_csv('laptopPrice.csv')

# 데이터 확인
print(df.head())  # 데이터프레임의 처음 5개 행 출력

# 막대 그래프(Bar Plot) 그리기
plt.bar(df['ram_gb'], df['Price'])
plt.xlabel('ram_gb')  # x축 레이블 설정
plt.ylabel('Price')  # y축 레이블 설정
plt.title('Bar Plot Example')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
# 업로드한 파일 읽기
df = pd.read_csv('laptopPrice.csv')

# 브랜드별 빈도 계산
brand_counts = df['brand'].value_counts()

# 브랜드별 비율 계산
brand_ratios = brand_counts / brand_counts.sum()

# 파이 차트 그리기
plt.pie(brand_ratios, labels=brand_ratios.index, autopct='%1.1f%%')
plt.title( "Laptop Brand Distribution")

# 그래프 표시
plt.show()
```
